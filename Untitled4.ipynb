```python
def prepare_data():
    """
    Prepares the training data and the test data, as a hdf5 file.
    Converts the RGB images to the 128 x 128 pixels.
    """
    shuffle_data = True
    hdf5_file = "C:\\Users\\lenovo\\horse-or-human.hdf5"
    train_horses = "C:\\Users\\lenovo\\horse-or-human\\train\horses\\*.png"
    train_humans = "C:\\Users\\lenovo\\horse-or-human\\train\\humans\\*.png"
    valid_horses = "C:\\Users\\lenovo\\horse-or-human\\validation\\horses\\*.png"
    valid_humans = "C:\\Users\\lenovo\\horse-or-human\\validation\\humans\\*.png"
    addrs_horses = glob.glob(train_horses)
    addrs_humans = glob.glob(train_humans)
    addrs_horses_val = glob.glob(valid_horses)
    addrs_humans_val = glob.glob(valid_humans)
    train_y = [0 for addr in addrs_horses]
    train_y.extend([1 for addr in addrs_humans])
    valid_y = [0 for addr in addrs_horses_val]
    valid_y.extend([1 for addr in addrs_humans_val])
    addrs_horses.extend(addrs_humans)
    addrs_horses_val.extend(addrs_humans_val)
    
    
    # Shuffles the training data and validation data
    if shuffle_data:
        train_bind = list(zip(addrs_horses, train_y))
        valid_bind = list(zip(addrs_horses_val, valid_y))
        shuffle(train_bind)
        shuffle(valid_bind)
        train_x, train_y = zip(*train_bind)
        valid_x, valid_y = zip(*valid_bind)
    
    # Create a hdf5 file and leave space for the images, create the dataset    
    train_shape = (len(train_x), 128, 128, 3)
    valid_shape = (len(valid_x), 128, 128, 3)
    f = h5py.File(hdf5_file, mode='w')
    f.create_dataset("train_x", train_shape, np.uint8)
    f.create_dataset("valid_x", valid_shape, np.uint8)
    f.create_dataset("train_y", (len(train_x),), np.uint8)
    f['train_y'][...] = train_y
    f.create_dataset("valid_y", (len(valid_x),), np.uint8)
    f['valid_y'][...] = valid_y
   
    
    # Resize and write the training images on the hdf5 file.
    for i in range(len(train_x)):
        if i % 100 == 0 and i > 1:
            print('Train data: {}/{}'.format(i, len(train_x)) )
        addr = train_x[i]
        img = cv2.imread(addr)
        img = cv2.resize(img, (128, 128), interpolation=cv2.INTER_CUBIC)# resize to (128,128)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # cv2 load images as BGR, convert it to RGB
        f["train_x"][i, ...] = img[None] 

    # Resize and write the test images on the hdf5 file.
    for i in range(len(valid_x)):
        if i % 100 == 0 and i > 1:
            print('Test data: {}/{}'.format(i, len(valid_x)) )
        addr = valid_x[i]
        img = cv2.imread(addr)
        img = cv2.resize(img, (128, 128), interpolation=cv2.INTER_CUBIC)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        f["valid_x"][i, ...] = img[None]

    f.close()
```

```python
def layer_unit_size(X, layer_size = 1, nn_class = 2):
    """
    Determine the neural network structure.
    """
    layers = []
    layers.append(X.shape[0])
    for i in range(0, layer_size - 1):
        layer = int(input("Layer size (" + str(i + 1)+ "): "))
        assert(layer > 0)
        layers.append(layer)
    if nn_class == 2:
        layers.append(1)
    else:
        layers.append(nn_class)
    return layers
```

```python
def initialize_parameters(layers):
    """
    Initialize the weights randomly using the Gaussian RV.
    Initialize the bias using the zeros.
    """
    np.random.seed(int(time.time()))
    scale = 0.01
    weight_bias = {}
    for layer in range(1, len(layers)):
        weight_bias['W' + str(layer)] = np.random.randn(layers[layer], layers[layer - 1]) * scale
        weight_bias['b' + str(layer)] = np.zeros((layers[layer], 1))
    assert(weight_bias['W' + str(layer)].shape == (layers[layer], layers[layer - 1]))
    assert(weight_bias['b' + str(layer)].shape == (layers[layer], 1))
    return weight_bias
```

```python
def linear_forward(activation, weight, bias):
    """
    Find the result of the net input function.
    Store the activation input, weight and bias.
    """
    linear_z = np.dot(weight, activation) + bias
    assert(linear_z.shape == (weight.shape[0], activation.shape[1]))
    cache = (activation, weight, bias)
    return linear_z, cache
```

```python
def sigmoid(linear_z):
    """
    Calculate the sigmoid activation function
    Return the activation output, activation input and the type of activation function.
    """
    return 1 / (1 + np.exp(-linear_z)), linear_z, "sigmoid"
```

```python
def relu(linear_z, leaky=False, leaky_rate=0):
    """
    Calculate the relu and leaky relu activation function
    Return the activation output, activation input and the type of activation function.
    """
    if leaky:
        return np.maximum(leaky_rate * linear_z, linear_z), linear_z, "leaky_relu"
    return np.maximum(leaky_rate * linear_z, linear_z), linear_z, "relu"
```

```python
def tanh(linear_z):
    """
    Calculate the tanh activation function
    Return the activation output, activation input and the type of activation function.
    """
    return np.tanh(linear_z), linear_z, "tanh"
```

```python
def softmax(linear_z):
    """
    Calculate the softmax activation function
    Return the activation output, activation input and the type of activation function.
    """
    new_max = np.exp(linear_z - np.max(linear_z, axis = 0, keepdims = True))
    return new_max / np.sum(new_max, axis = 0, keepdims = True), linear_z, "softmax"
```

```python
def activation_forward(activation_prev, weight, bias, activation_type):
    """
    Calculate the activation output
    Store the activation output and activation input in the cache.
    """
    linear_z, linear_cache = linear_forward(activation_prev, weight, bias)
    if activation_type == "relu":
        activation, activation_cache, activation_type = relu(linear_z)
    elif activation_type == "leaky_relu":
        activation, activation_cache, activation_type = relu(linear_z, True, 0.01)
    elif activation_type == "sigmoid":
        activation, activation_cache, activation_type = sigmoid(linear_z)
    elif activation_type == "tanh":
        activation, activation_cache, activation_type = tanh(linear_z)
    elif activation_type == "softmax":
        activation, activation_cache, activation_type = softmax(linear_z)
    assert(activation.shape == (weight.shape[0], activation_prev.shape[1]))
    cache = (linear_cache, activation_cache)
    return activation, cache
```

```python
def forward_model(X, weight_bias, prev_activation_types, nn_class = 2):
    """
    Simulate the forward propagation
    Store the activation output and activation input and activation type in the collective cache.
    """
    assert(nn_class > 1)
    caches = []
    activation = X
    length = len(weight_bias) // 2
    if len(prev_activation_types) == 0:
        activation_types = []
        print("The choice of activation types you can choose are relu, leaky_relu, sigmoid or tanh.")
        for lno in range(1, length):
            activation_prev = activation
            activation_type = input("Enter the activation type for layer (" + str(lno) + "): ")
            assert(activation_type == "relu" or activation_type == "leaky_relu" or activation_type == "sigmoid" or activation_type == "tanh")
            activation, cache = activation_forward(activation_prev, weight_bias['W'+str(lno)], weight_bias['b'+str(lno)], activation_type)
            caches.append(cache)
            activation_types.append(activation_type)
        if nn_class == 2:
            print("The choice of activation types you can choose are sigmoid or tanh.")
            activation_type = input("Enter the activation type for layer (" + str(lno) + "): ")
            assert(activation_type == "sigmoid" or activation_type == "tanh")
            activation_layer, cache = activation_forward(activation, weight_bias['W'+str(length)], weight_bias['b'+str(length)], activation_type)
            activation_types.append(activation_type)
        else:
            activation_layer, cache = activation_forward(activation, weight_bias['W'+str(length)], weight_bias['b'+str(length)], "softmax")
            activation_types.append("softmax")
        caches.append(cache)
    else:
        activation_types = prev_activation_types
        for lno in range(1, length):
            activation_prev = activation
            activation, cache = activation_forward(activation_prev, weight_bias['W'+str(lno)], weight_bias['b'+str(lno)], prev_activation_types[lno - 1])
            caches.append(cache)
        activation_layer, cache = activation_forward(activation, weight_bias['W'+str(length)], weight_bias['b'+str(length)], prev_activation_types[lno])
        caches.append(cache)
    assert(activation_layer.shape == (1, X.shape[1]))
    return activation_layer, caches, activation_types
```

```python
def compute_cost(activation_layer, Y, last_type):
    """
    Compute the cost of output according to the ground truth.
    """
    if last_type == "tanh":
        cost = -np.sum(-0.5*((1 + Y) * np.log(activation_layer + 1 + 1e-15) + (1 - Y) * np.log(1 + 1e-15 - activation_layer)) + np.log(2)) / Y.shape[1]
    else:
        cost = -np.sum(Y * np.log((1 + activation_layer) / 2 + 1e-15) + (1 - Y) * np.log((1 - activation_layer) / 2 + 1e-15)) / Y.shape[1]
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    return cost
```

```python
def linear_backward(dz_linear, cache):
    """
    Calculate the change of weight, bias and activation with respect to the cost function.
    """
    activation_prev, weight, bias = cache
    train_size =  activation_prev.shape[1]
    dweight = np.dot(dz_linear, activation_prev.T) / train_size
    dbias = np.sum(dz_linear, axis = 1, keepdims = True) / train_size
    dactivation_prev = np.dot(weight.T, dz_linear)
    assert(dactivation_prev.shape == activation_prev.shape)
    assert(dweight.shape == weight.shape)
    assert(dbias.shape == bias.shape)
    return dactivation_prev, dweight, dbias
```

```python
def sigmoid_backward(dactivation, cache):
    """
    Calculate the sigmoid activation backpropogation function
    Return the change of activation.
    """
    sigmoid_cache = sigmoid(cache)[0]
    return dactivation * sigmoid_cache * (1 - sigmoid_cache) 
```

```python
def relu_backward(dactivation, cache, leaky=False, leaky_rate=0):
    """
    Calculate the relu and leaky relu activation backpropogation function
    Return the change of activation.
    """
    if leaky:
        relu_cache = np.where(cache < 0, leaky_rate, 1)
    else:
        relu_cache = np.where(cache < 0, 0, 1)
    return dactivation * relu_cache
```

```python
def tanh_backward(dactivation, cache):
    """
    Calculate the tanh activation backpropogation function
    Return the change of activation.
    """
    tanh_cache = tanh(cache)[0]
    return dactivation * (1 - tanh_cache ** 2) 
```

```python
def softmax_backward(dactivation, cache):
    """
    Calculate the softmax activation backpropogation function
    Return the change of activation.
    """
    softmax_cache = softmax(cache)[0]
    softmax_diag = np.diag(softmax_cache) 
    soft_vector = softmax_diag.reshape(softmax_diag.shape[0], 1)
    soft_matrix = np.tile(soft_vector, softmax_cache.shape[0])
    softmax_cache = softmax_diag - soft_matrix * soft_matrix.T
    return dactivation * softmax_cache
```

```python
def linear_activation_backward(dactivation, cache, activation_type, leaky=False, leaky_rate=0):
    """
    Calculate the activation output
    Store the activation output and activation input in the cache.
    """
    linear_cache, activation_cache = cache
    if activation_type == "relu":
        dz = relu_backward(dactivation, activation_cache)
    elif activation_type == "leaky_relu":
        dz = relu_backward(dactivation, activation_cache, True, 0.01)
    elif activation_type == "sigmoid":
        dz = sigmoid_backward(dactivation, activation_cache)
    elif activation_type == "tanh":
        dz = tanh_backward(dactivation, activation_cache)
    elif activation_type == "softmax":
        dz = softmax_backward(dactivation, activation_cache)
    dactivation_prev, dweight, dbias = linear_backward(dz, linear_cache)
    return dactivation_prev, dweight, dbias
```

```python
def linear_model_backward(activation, Y, caches, activation_types):
    """
    Simulate the back propagation
    Find the gradients with respect to the cost function.
    """
    grads = {}
    length = len(caches)
    dlast_activation = - (np.divide(Y, activation) - np.divide(1 - Y, 1 - activation))
    current_cache = caches[length - 1]
    grads["dA" + str(length - 1)], grads["dW" + str(length)], grads["db" + str(length)] = linear_activation_backward(dlast_activation, current_cache, activation_types[-1])
    for layer in reversed(range(length - 1)):
        current_cache = caches[layer]
        if activation_types[layer] == "leaky_relu":
            dactivation_prev_temp, dweight_prev_temp, dbias_prev_temp = linear_activation_backward(grads["dA" + str(layer + 1)], current_cache, activation_types[layer], True, 0.01)
        else:
            dactivation_prev_temp, dweight_prev_temp, dbias_prev_temp = linear_activation_backward(grads["dA" + str(layer + 1)], current_cache, activation_types[layer])
        grads["dA" + str(layer)], grads["dW" + str(layer + 1)], grads["db" + str(layer + 1)] = dactivation_prev_temp, dweight_prev_temp, dbias_prev_temp
    return grads
```

```python
def update_parameters(parameters, grads, learning_rate):
    """
    Update the parameters
    Return the updated parameters
    """
    length = len(parameters) // 2
    for layer in range(length):
        parameters["W" + str(layer + 1)] = parameters["W" + str(layer + 1)] - learning_rate * grads["dW" + str(layer + 1)]
        parameters["b" + str(layer + 1)] = parameters["b" + str(layer + 1)] - learning_rate * grads["db" + str(layer + 1)]
    return parameters
```

```python
def multi_layer_model(X, Y, nn_class=2, learning_rate=0.001, num_iterations=3000, print_cost=False):
    """
    Create a multi layer model using the prebuild functions
    Do forward propagation and back propogation and return the parameters
    """
    layer_size = int(input("How many layers do you need? :"))
    layers = layer_unit_size(X, layer_size, nn_class)
    costs = []
    parameters = initialize_parameters(layers)
    activation_types = []
    for i in range(0, num_iterations):
        A_Layer, caches, activation_types = forward_model(X, parameters, activation_types, nn_class)
        cost = compute_cost(A_Layer, Y, activation_types[-1])
        grads = linear_model_backward(A_Layer, Y, caches, activation_types)
        parameters = update_parameters(parameters, grads, learning_rate)
        if print_cost and i % 100 == 0:
            print('Cost after iteration {}:{}'.format(i, cost))
            costs.append(cost)
    plt.plot(np.squeeze(costs))
    plt.ylabel("Costs")
    plt.xlabel("Number of Iterations (per 100)")
    plt.title("Learning Rate: " + str(learning_rate))
    plt.show()
    return parameters, activation_types
```

```python
def predict(X, Y, weight_bias, prev_activation_types, nn_class = 2, data="Train"):
    """
    Predict the result using validation data.
    Find the accuracy of the data.
    """
    A_Layer, caches, activation_types = forward_model(X, weight_bias, prev_activation_types, nn_class)
    A_Layer = np.where(A_Layer >= np.sum(A_Layer) / A_Layer.shape[1], 1, 0)
    predicted = (1 - np.count_nonzero(A_Layer - Y) / A_Layer.shape[1]) * 100
    print(data + " Accuracy: " + str(predicted))
    return predicted
```
